#📌 Extracción

```python
import pandas as pd
import numpy as np
import matplotlib.pyplot as plt
import seaborn as sns
import plotly.express as px
```

Creacion del DataFrame

```python
url = ('https://raw.githubusercontent.com/alura-cursos/challenge2-data-science-LATAM/refs/heads/main/TelecomX_Data.json')
```

```python
data = pd.read_json(url)
df = pd.DataFrame(data)
```

```python
normal = df.columns[2:].tolist()


for col in normal:
    norm = pd.json_normalize(df[col])
    print(norm.shape)
    df.drop(col, axis=1, inplace=True)
    df = df.join(norm)
```

#📊 Carga y análisis

```python
df['Churn'] = df['Churn'].map({0: 'Se quedaron', 1: 'Se fueron'})
churn_counts = df['Churn'].value_counts().reset_index()
churn_counts.columns = ['Churn', 'Cantidad']


fig = px.pie(churn_counts,
    names='Churn',
    values='Cantidad',
    title='Distribución de Churn (Clientes que se fueron vs. se quedaron)',
    color='Churn',
    labels = churn_counts.index,
    color_discrete_map={0: 'lightgreen', 1: 'tomato'},
    width = 600, height = 600
)
fig.show()
```

```python
def churn_por_variable(df, columna, ancho=700, alto=450):
    if df['Churn'].dtype != 'O':
        df['Churn'] = df['Churn'].map({0: 'Se quedaron', 1: 'Se fueron'})
    resumen = df.groupby([columna, 'Churn']).size().reset_index(name='Cantidad')


    total_por_categoria = resumen.groupby(columna)['Cantidad'].transform('sum')
    resumen['Porcentaje'] = resumen['Cantidad'] / total_por_categoria * 100


    fig = px.bar(
        resumen,
        x=columna,
        y='Porcentaje',
        color='Churn',
        barmode='stack',
        title=f'Proporción de Churn según {columna}',
        labels={'Porcentaje': 'Porcentaje (%)'},
        color_discrete_map={'Se quedaron': 'lightgreen', 'Se fueron': 'tomato'},
        width=ancho,
        height=alto
    )



    fig.show()
```

```python
churn_por_variable(df, 'Genero')
churn_por_variable(df, 'Contrato')
churn_por_variable(df, 'Metodo_pago')
```

```python
def comparar_variable_numerica(df, variable, ancho=700, alto=450):

    if df['Churn'].dtype != 'O':
        df['Churn'] = df['Churn'].map({0: 'Se quedaron', 1: 'Se fueron'})


    fig = px.box(
        df,
        x='Churn',
        y=variable,
        color='Churn',
        title=f'Distribución de {variable} según Churn',
        color_discrete_map={'Se quedaron': 'lightgreen', 'Se fueron': 'tomato'},
        width=ancho,
        height=alto,
        points='all'
    )

    fig.show()

comparar_variable_numerica(df, 'Charges.Total')
comparar_variable_numerica(df, 'Factura_mensual')
comparar_variable_numerica(df, 'Meses_contrato')
```
